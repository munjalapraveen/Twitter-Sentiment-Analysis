# Data Cleaning

```python
def Data_Clean(Tweet):
    Tweet_Pun_Rm=[word for word in Tweet if word not in string.punctuation]
    Tweet_Pun_Rm_Join=''.join(Tweet_Pun_Rm)
    Tweet_Stop=[st for st in Tweet_Pun_Rm_Join.split() if st.lower() not in stopwords.words('english')]
    Tweet_Stop_join=' '.join(Tweet_Stop)
    return Tweet_Stop_join
```

```python
twitter_df_clean=twitter_df['tweet'].apply(Data_Clean)
```

```python
vector=CountVectorizer(analyzer=Data_Clean).fit_transform(twitter_df['tweet']).toarray()
```

```python
X=vector
y=twitter_df['label']
```

# Building Model

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
naviesclassifier=MultinomialNB()
```

```python
naviesclassifier.fit(X_train,y_train)
```

```python
y_pred= naviesclassifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
acc=(cm[0][0]+cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0])*100
```
